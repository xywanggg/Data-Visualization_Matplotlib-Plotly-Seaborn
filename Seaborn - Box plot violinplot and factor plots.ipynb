# box plot violinplot and factor plots

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
housing_data=pd.read_csv('housing.csv')
housing_data.head(5)
```

```python
f,ax=plt.subplots(figsize=(12,8))
sns.boxplot(x='ocean_proximity',y='median_house_value',data=housing_data)
```

```python
median=housing_data['median_house_value'].median()
median
```

```python
housing_data['above_median']=(housing_data['median_house_value']-median)>0
housing_data.sample(5)
```

```python
f,ax=plt.subplots(figsize=(12,8))
sns.boxplot(x='ocean_proximity',y='median_income',hue='above_median',data=housing_data)
```

```python
f,ax=plt.subplots(figsize=(12,8))
sns.violinplot(x='ocean_proximity',y='median_house_value',data=housing_data)
```

```python
mpg_data=pd.read_csv('auto-mpg.csv')
mpg_data.replace('.',np.nan,inplace=True)
mpg_data.replace('?',np.nan,inplace=True)
mpg_data=mpg_data.dropna()
mpg_data.head()
```

```python
f,ax=plt.subplots(figsize=(12,8))
sns.violinplot(x='model year',y='mpg',data=mpg_data)
```

```python
f,ax=plt.subplots(figsize=(12,8))
sns.barplot(x='cylinders',y='mpg',data=mpg_data)
```

```python
f,ax=plt.subplots(figsize=(12,8))
sns.pointplot(x='cylinders',y='mpg',data=mpg_data)
```

```python
#frequency
sns.catplot(x='cylinders',
            data=mpg_data,height=5,aspect=2,
           kind='count')
```
