# strip plots and swarm plots

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
mpg_data=pd.read_csv('auto-mpg.csv')
mpg_data.head()
```

```python
#replace missing data
mpg_data.replace('.',np.nan,inplace=True)
mpg_data.replace('?',np.nan,inplace=True)
mpg_data=mpg_data.dropna()
mpg_data.head()
```

```python
mpg_data['horsepower']=mpg_data['horsepower'].astype(int)
```

```python
plt.figure(figsize=(12,8))
sns.stripplot(mpg_data['mpg'])
```

```python
plt.figure(figsize=(12,8))
sns.stripplot(x='cylinders',y='mpg',
             data=mpg_data,jitter=True)
```

```python
plt.figure(figsize=(12,8))
sns.stripplot(x='cylinders',y='mpg',hue='origin',
             data=mpg_data,jitter=True)
```

```python
plt.figure(figsize=(12,8))
sns.swarmplot(x='cylinders',y='mpg',
             data=mpg_data)
```

```python
plt.figure(figsize=(12,8))
sns.swarmplot(x='cylinders',y='mpg',hue='origin',dodge=True,
             data=mpg_data)
```
