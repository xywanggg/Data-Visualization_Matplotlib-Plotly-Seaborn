```python
import pandas as pd
import plotly
import plotly.graph_objs as go
import plotly.offline as offline
```

```python
offline.init_notebook_mode(connected=True)
```

# bubble charts

```python
trace=go.Scatter(x=[15,18,21,25],
                y=[100,400,300,200],
                mode='markers')

data=[trace]

offline.iplot(data)
```

```python
z=[25,100,75,50]
```

```python
trace=go.Scatter(x=[15,18,21,25],
                y=[100,400,300,200],
                mode='markers',
                marker=dict(size=z))

data=[trace]

offline.iplot(data)
```

```python
i=[5,6,8,4]
```

```python
trace=go.Scatter(x=[15,18,21,25],
                y=[100,400,300,200],
                mode='markers',
                marker=dict(size=z,
                           color=i,
                           colorscale='Portland',
                           showscale=True))
```

```python
data=[trace]

offline.iplot(data)
```

```python
housing_data=pd.read_csv('housing.csv')
housing_data.sample(10)
```

```python
housing_data.shape
```

```python
housing_data=housing_data.sample(frac=0.07).reset_index(drop=True)
housing_data.shape
```

```python
trace=go.Scatter(x=housing_data['median_income'],
                y=housing_data['median_house_value'],
                mode='markers',
                marker=dict(size=housing_data['total_rooms'],
                           sizeref=500,
                           color=housing_data['housing_median_age'],
                           colorscale='Jet',
                           showscale=True))
```

```python
data=[trace]

layout=go.Layout(height=600,width=900,
                title='Housing Data',
                hovermode='closest')

fig=go.Figure(data=data,
             layout=layout)
offline.iplot(fig)
```

# bubble maps

```python
trace=dict(type='scattergeo',
          lon=[-97.92,0,78.8],
          lat=[39.3,0,21.76],
          marker=dict(size=10),
          mode='markers',)
```

```python
data=[trace]
layout=dict(showlegend=False,
           geo=dict(showland=True))
```

```python
fig=dict(data=data,
        layout=layout)
offline.iplot(fig)
```

```python
housing_data=pd.read_csv('housing.csv')
housing_data.sample(10)
```

```python
housing_data=housing_data.sample(frac=0.1).reset_index(drop=True)
housing_data.shape
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
housing_data['ocean_proximity_labels']=le.fit_transform(housing_data['ocean_proximity'])
```

```python
housing_data[['ocean_proximity','ocean_proximity_labels']].sample(5)
```

```python
trace=dict(type='scattergeo',
          lat=housing_data['latitude'],
          lon=housing_data['longitude'],
          marker=dict(size=housing_data['median_house_value']/1000,
                     sizemode='area',
                     color=housing_data['ocean_proximity_labels'],
                     colorscale='Portland',
                     showscale=True),
          mode='markers')
```

```python
data=[trace]
layout=dict(showlegend=False,
           geo=dict(showland=True,
                   landcolor='yellow'))
```

```python
fig=dict(data=data,layout=layout)
offline.iplot(fig)
```
