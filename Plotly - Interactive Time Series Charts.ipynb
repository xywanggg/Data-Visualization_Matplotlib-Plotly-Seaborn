# Interactive Time series charts

```python
import pandas as pd
import plotly.graph_objs as go
import plotly.offline as offline

offline.init_notebook_mode(connected=True)
```

```python
from IPython.display import Image
```

```python
stock_data=pd.read_csv('stock_market.csv')
stock_data.sample(5)
```

```python
stock_data.shape
```

```python
trace_close=go.Scatter(x=stock_data.date,
                      y=stock_data.close,
                      name='Stock Close')
data=[trace_close]
offline.iplot(data)
```

```python
Image('newplot.png')
```

```python
trace_open=go.Scatter(x=stock_data.date,
                      y=stock_data.open,
                      name='Stock Open',
                      line=dict(color='orange'))
data=[trace_close]
offline.iplot(data)
```

```python
Image('newplot2.png')
```

```python
data=[trace_close,trace_open]
layout=dict(title='Stock Price Data')
```

```python
fig=dict(data=data,
        layout=layout)
offline.iplot(fig)
```

```python
Image('newplot3.png')
```

```python
layout=dict(title='Stock Price Data',
           xaxis=dict(rangeslider=dict(visible=True),
                     type='date'))
fig=dict(data=data,
        layout=layout)
offline.iplot(fig)
```

```python
Image('newplot4.png')
```

# candlestick charts

```python
trace=go.Candlestick(x=stock_data.date,
                    open=stock_data.open,
                    high=stock_data.high,
                    low=stock_data.low,
                    close=stock_data.close)
```

```python
data=[trace]
offline.iplot(data)
```

```python
Image('newplot5.png')
```

```python
trace=go.Candlestick(x=stock_data.date[0:10],
                    open=stock_data.open[0:10],
                    high=stock_data.high[0:10],
                    low=stock_data.low[0:10],
                    close=stock_data.close[0:10])
```

```python
data=[trace]
offline.iplot(data)
```
