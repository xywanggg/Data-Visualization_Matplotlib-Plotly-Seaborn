# univariate and bivariate distributions

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
weight_data=pd.read_csv('weight-height.csv')
weight_data.sample(10)
```

```python
weight_data.shape
```

```python
plt.figure(figsize=(12,8))
sns.distplot(weight_data['Weight'],color='purple')
plt.title('Weights')
plt.ylabel('Count')
```

```python
plt.figure(figsize=(12,8))
sns.distplot(weight_data['Weight'],color='purple',
            bins=70)

plt.title('Weights')
plt.ylabel('Density')
```

```python
plt.figure(figsize=(12,8))
sns.distplot(weight_data['Weight'],hist=False,color='purple')

plt.title('Weights')
plt.ylabel('Density')
```

```python
plt.figure(figsize=(12,8))
sns.distplot(weight_data['Weight'],kde=False,
             rug=True,color='purple',)

plt.title('Weights')
plt.ylabel('Density')
```

```python
plt.figure(figsize=(12,8))
sns.rugplot(weight_data['Weight'],height=0.5,
             axis='x',color='purple',)

plt.title('Weights')
plt.ylabel('Density')
```

```python
plt.figure(figsize=(12,8))
sns.set(color_codes=True)
sns.kdeplot(weight_data['Weight'],shade=True,
            color='purple')

plt.title('Weights')
plt.ylabel('Density')
```

```python
mpg_data=pd.read_csv('auto-mpg.csv')
mpg_data.head()
```

```python
#replace missing data
mpg_data.replace('.',np.nan,inplace=True)
mpg_data.replace('?',np.nan,inplace=True)
mpg_data=mpg_data.dropna()
mpg_data.head()
```

```python
mpg_data['horsepower']=mpg_data['horsepower'].astype(int)
```

```python
mpg_data.shape
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='horsepower',y='mpg',data=mpg_data)
plt.xlabel('horsepower')
plt.ylabel('mpg')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='horsepower',y='mpg',data=mpg_data,hue='cylinders')
plt.xlabel('horsepower')
plt.ylabel('mpg')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='horsepower',y='mpg',
                hue='cylinders',
                size='weight',
                sizes=(30,300),hue_norm=(0,9),
                data=mpg_data)
plt.xlabel('horsepower')
plt.ylabel('mpg')
plt.show()
```

```python
sns.jointplot(x='horsepower',y='mpg',data=mpg_data)
```

```python
with sns.axes_style('white'):
    sns.jointplot(x='horsepower',y='mpg',
                 data=mpg_data,kind='hex',
                 color='k')
```
