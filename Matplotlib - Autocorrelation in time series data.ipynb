```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# autocorrelation

```python
grand_canyon_data=pd.read_csv('park_visits.csv')
grand_canyon_data.head()
```

```python
grand_canyon_data['NumVisits'].describe()
```

```python
grand_canyon_data['NumVisits']=grand_canyon_data['NumVisits']/1000
grand_canyon_data['NumVisits'].describe()
```

```python
plt.figure(figsize=(12,8))
plt.acorr(grand_canyon_data['NumVisits'],
          maxlags=6)
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.acorr(grand_canyon_data['NumVisits'],
          maxlags=13,color='green')
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.acorr(grand_canyon_data['NumVisits'],
          maxlags=48,color='green')
plt.show()
```

# stacked plot and stem plot

```python
np_data=pd.read_csv('national_park.csv')
np_data.head()
```

```python
x=np_data['Year']
```

```python
y=np.vstack([np_data['Badlands'],
            np_data['GrandCanyon'],
            np_data['BryceCanyon']])
y
```

```python
labels=['Badlands',
       'GrandCanyon',
       'BryceCanyon']
```

```python
plt.figure(figsize=(12,8))
plt.stackplot(x,y,labels=labels)
plt.legend(loc='upper left')
plt.show()
```

```python
colors=['sandybrown','tomato','skyblue']
```

```python
plt.figure(figsize=(12,8))

plt.stackplot(x,y,labels=labels,
             colors=colors,
             edgecolor='grey')

plt.legend(loc='upper left')
plt.show()
```

```python
np_data[['Badlands',
         'GrandCanyon',
         'BryceCanyon']]=np_data[['Badlands',
                                 'GrandCanyon',
                                 'BryceCanyon']].diff()
np_data.head()
```

```python
plt.figure(figsize=(12,8))
plt.stem(np_data['Year'],
        np_data['GrandCanyon'],
        use_line_collection=True)

plt.title('change in number of visitors')

plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.stem(np_data['Year'],
        np_data['BryceCanyon'],
         markerfmt='g_',
         linefmt='r--',
         basefmt='b:',
        use_line_collection=True)

plt.title('change in number of visitors')

plt.show()
```
